```python
"""
Complete ISIC 2018 Skin Lesion Segmentation Pipeline with DeepLabV3+
End-to-End Implementation: Data Loading → Training → Evaluation → Inference
"""

import os
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
from tqdm import tqdm
import json
import albumentations as A
from albumentations.pytorch import ToTensorV2

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import torch.optim as optim
from torchvision.models import resnet50, resnet101

import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
def set_seed(seed=42):
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

# ============================================================================
# 1. MODEL ARCHITECTURE - DeepLabV3+
# ============================================================================

class ASPP(nn.Module):
    """Atrous Spatial Pyramid Pooling Module"""
    def __init__(self, in_channels, out_channels=256):
        super(ASPP, self).__init__()

        # Different atrous rates for multi-scale feature extraction
        atrous_rates = [6, 12, 18]

        # 1x1 convolution
        self.conv1 = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, 1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

        # Atrous convolutions with different rates
        self.conv2 = self._make_atrous_conv(in_channels, out_channels, atrous_rates[0])
        self.conv3 = self._make_atrous_conv(in_channels, out_channels, atrous_rates[1])
        self.conv4 = self._make_atrous_conv(in_channels, out_channels, atrous_rates[2])

        # Global average pooling branch
        self.global_avg_pool = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(in_channels, out_channels, 1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

        # Fusion of all branches
        self.conv_fusion = nn.Sequential(
            nn.Conv2d(out_channels * 5, out_channels, 1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5)
        )

    def _make_atrous_conv(self, in_channels, out_channels, dilation):
        return nn.Sequential(
            nn.Conv2d(in_channels, out_channels, 3, padding=dilation,
                     dilation=dilation, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        size = x.shape[2:]

        # Apply all branches
        x1 = self.conv1(x)
        x2 = self.conv2(x)
        x3 = self.conv3(x)
        x4 = self.conv4(x)
        x5 = self.global_avg_pool(x)
        x5 = F.interpolate(x5, size=size, mode='bilinear', align_corners=True)

        # Concatenate all features
        x = torch.cat([x1, x2, x3, x4, x5], dim=1)

        # Fuse features
        return self.conv_fusion(x)


class Decoder(nn.Module):
    """DeepLabV3+ Decoder Module"""
    def __init__(self, low_level_channels, num_classes):
        super(Decoder, self).__init__()

        # Reduce low-level feature channels
        self.conv_low_level = nn.Sequential(
            nn.Conv2d(low_level_channels, 48, 1, bias=False),
            nn.BatchNorm2d(48),
            nn.ReLU(inplace=True)
        )

        # Decoder convolutions with depthwise separable convolutions
        self.conv_decoder = nn.Sequential(
            # First depthwise separable conv
            nn.Conv2d(304, 304, 3, padding=1, groups=304, bias=False),  # Depthwise
            nn.BatchNorm2d(304),
            nn.ReLU(inplace=True),
            nn.Conv2d(304, 256, 1, bias=False),  # Pointwise
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),

            # Second depthwise separable conv
            nn.Conv2d(256, 256, 3, padding=1, groups=256, bias=False),  # Depthwise
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, 1, bias=False),  # Pointwise
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Dropout(0.1)
        )

        # Final classification layer
        self.classifier = nn.Conv2d(256, num_classes, 1)

    def forward(self, x, low_level_feat):
        # Process low-level features
        low_level_feat = self.conv_low_level(low_level_feat)

        # Upsample high-level features to match low-level feature size
        x = F.interpolate(x, size=low_level_feat.shape[2:],
                         mode='bilinear', align_corners=True)

        # Concatenate features
        x = torch.cat([x, low_level_feat], dim=1)

        # Apply decoder convolutions
        x = self.conv_decoder(x)

        # Final classification
        return self.classifier(x)


class DeepLabV3Plus(nn.Module):
    """DeepLabV3+ for ISIC 2018 Skin Lesion Segmentation"""
    def __init__(self, num_classes=1, backbone='resnet50',
                 pretrained=True, output_stride=16):
        super(DeepLabV3Plus, self).__init__()

        self.num_classes = num_classes
        self.output_stride = output_stride

        # Load backbone
        if backbone == 'resnet50':
            resnet = resnet50(pretrained=pretrained)
        elif backbone == 'resnet101':
            resnet = resnet101(pretrained=pretrained)
        else:
            raise ValueError(f"Unsupported backbone: {backbone}")

        # Encoder - Extract features from ResNet
        self.conv1 = resnet.conv1
        self.bn1 = resnet.bn1
        self.relu = resnet.relu
        self.maxpool = resnet.maxpool

        self.layer1 = resnet.layer1  # Low-level features (256 channels)
        self.layer2 = resnet.layer2
        self.layer3 = resnet.layer3
        self.layer4 = resnet.layer4  # High-level features (2048 channels)

        # Modify layer4 for atrous convolutions if output_stride=16
        if output_stride == 16:
            self._modify_resnet_layer(self.layer4, dilation=2)
        elif output_stride == 8:
            self._modify_resnet_layer(self.layer3, dilation=2)
            self._modify_resnet_layer(self.layer4, dilation=4)

        # ASPP module
        self.aspp = ASPP(in_channels=2048, out_channels=256)

        # Decoder module
        self.decoder = Decoder(low_level_channels=256, num_classes=num_classes)

        self._init_weights()

    def _modify_resnet_layer(self, layer, dilation):
        """Modify ResNet layer to use atrous convolutions"""
        for block in layer:
            block.conv2.dilation = (dilation, dilation)
            block.conv2.padding = (dilation, dilation)

    def _init_weights(self):
        """Initialize decoder weights"""
        for m in self.decoder.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        input_size = x.shape[2:]

        # Encoder - Extract features
        x = self.conv1(x)
        x = self.bn1(x)
        x = self.relu(x)
        x = self.maxpool(x)

        # Low-level features (after layer1)
        low_level_feat = self.layer1(x)

        # High-level features
        x = self.layer2(low_level_feat)
        x = self.layer3(x)
        x = self.layer4(x)

        # ASPP for multi-scale feature extraction
        x = self.aspp(x)

        # Decoder to recover spatial information
        x = self.decoder(x, low_level_feat)

        # Upsample to input size
        x = F.interpolate(x, size=input_size, mode='bilinear', align_corners=True)

        return x


# ============================================================================
# 2. LOSS FUNCTIONS
# ============================================================================

class DiceLoss(nn.Module):
    """Dice Loss for segmentation"""
    def __init__(self, smooth=1.0):
        super(DiceLoss, self).__init__()
        self.smooth = smooth

    def forward(self, pred, target):
        pred = torch.sigmoid(pred)
        pred = pred.view(-1)
        target = target.view(-1)

        intersection = (pred * target).sum()
        dice = (2. * intersection + self.smooth) / (pred.sum() + target.sum() + self.smooth)

        return 1 - dice


class FocalLoss(nn.Module):
    """Focal Loss for handling class imbalance"""
    def __init__(self, alpha=0.25, gamma=2.0):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, pred, target):
        bce_loss = F.binary_cross_entropy_with_logits(pred, target, reduction='none')
        pt = torch.exp(-bce_loss)
        focal_loss = self.alpha * (1 - pt) ** self.gamma * bce_loss
        return focal_loss.mean()


class CombinedLoss(nn.Module):
    """Combined Dice + Focal Loss as per project spec (0.7 Dice + 0.3 Focal)"""
    def __init__(self, dice_weight=0.7, focal_weight=0.3, focal_alpha=0.25, focal_gamma=2.0):
        super(CombinedLoss, self).__init__()
        self.dice_weight = dice_weight
        self.focal_weight = focal_weight
        self.dice_loss = DiceLoss()
        self.focal_loss = FocalLoss(alpha=focal_alpha, gamma=focal_gamma)

    def forward(self, pred, target):
        dice = self.dice_loss(pred, target)
        focal = self.focal_loss(pred, target)
        return self.dice_weight * dice + self.focal_weight * focal


# ============================================================================
# 3. METRICS
# ============================================================================

def calculate_iou(pred, target, threshold=0.5):
    """Jaccard Index (IoU)"""
    pred = (pred > threshold).float()
    intersection = (pred * target).sum()
    union = pred.sum() + target.sum() - intersection
    iou = (intersection + 1e-6) / (union + 1e-6)
    return iou.item()


def calculate_dice(pred, target, threshold=0.5):
    """Dice Similarity Coefficient"""
    pred = (pred > threshold).float()
    intersection = (pred * target).sum()
    dice = (2. * intersection + 1e-6) / (pred.sum() + target.sum() + 1e-6)
    return dice.item()


def calculate_metrics(pred, target, threshold=0.5):
    """Calculate all metrics"""
    pred_binary = (pred > threshold).float()
    target_binary = target.float()

    # Flatten tensors
    pred_flat = pred_binary.view(-1)
    target_flat = target_binary.view(-1)

    # True/False Positives/Negatives
    tp = (pred_flat * target_flat).sum()
    fp = (pred_flat * (1 - target_flat)).sum()
    fn = ((1 - pred_flat) * target_flat).sum()
    tn = ((1 - pred_flat) * (1 - target_flat)).sum()

    # Metrics
    iou = calculate_iou(pred, target, threshold)
    dice = calculate_dice(pred, target, threshold)
    sensitivity = (tp + 1e-6) / (tp + fn + 1e-6)
    specificity = (tn + 1e-6) / (tn + fp + 1e-6)
    precision = (tp + 1e-6) / (tp + fp + 1e-6)

    return {
        'iou': iou,
        'dice': dice,
        'sensitivity': sensitivity.item(),
        'specificity': specificity.item(),
        'precision': precision.item()
    }


# ============================================================================
# 4. DATASET AND AUGMENTATION
# ============================================================================

class ISICDataset(Dataset):
    """ISIC 2018 Dataset"""
    def __init__(self, image_dir, mask_dir, image_list, transform=None):
        self.image_dir = image_dir
        self.mask_dir = mask_dir
        self.image_list = image_list
        self.transform = transform

    def __len__(self):
        return len(self.image_list)

    def __getitem__(self, idx):
        img_name = self.image_list[idx]

        # Load image
        img_path = os.path.join(self.image_dir, img_name + '.jpg')
        image = np.array(Image.open(img_path).convert('RGB'))

        # Load mask
        mask_path = os.path.join(self.mask_dir, img_name + '_segmentation.png')
        mask = np.array(Image.open(mask_path).convert('L'))
        mask = (mask > 127).astype(np.float32)  # Binary threshold

        # Apply augmentations
        if self.transform:
            augmented = self.transform(image=image, mask=mask)
            image = augmented['image']
            mask = augmented['mask']

        mask = mask.unsqueeze(0)  # Add channel dimension
        return image, mask


def get_transforms(phase='train'):
    """Get augmentation transforms - compatible with albumentations >= 2.x"""
    if phase == 'train':
        return A.Compose([
            A.Resize(height=512, width=512),
            A.HorizontalFlip(p=0.5),
            A.VerticalFlip(p=0.5),
            A.Rotate(limit=45, p=0.5),
            A.ElasticTransform(alpha=50, sigma=10, p=0.3),
            A.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1, p=0.5),
            A.GaussianBlur(blur_limit=(3, 7), p=0.3),

            # 🔴 CHANGED LINE: use `size=(512, 512)` instead of separate h, w
            A.RandomResizedCrop(size=(512, 512), scale=(0.8, 1.0), p=0.5),

            A.Normalize(mean=[0.485, 0.456, 0.406],
                        std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
    else:
        return A.Compose([
            A.Resize(height=512, width=512),
            A.Normalize(mean=[0.485, 0.456, 0.406],
                        std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])



def prepare_data_splits(image_dir, train_split=0.8, val_split=0.1):
    """Split data into train/val/test sets (80/10/10 as per spec)"""
    # Get all image names (without extensions)
    all_images = [f.replace('.jpg', '') for f in os.listdir(image_dir) if f.endswith('.jpg')]
    all_images = sorted(all_images)

    # Shuffle
    np.random.shuffle(all_images)

    # Split indices
    n_total = len(all_images)
    n_train = int(n_total * train_split)
    n_val = int(n_total * val_split)

    train_list = all_images[:n_train]
    val_list = all_images[n_train:n_train + n_val]
    test_list = all_images[n_train + n_val:]

    print(f"Data split - Train: {len(train_list)}, Val: {len(val_list)}, Test: {len(test_list)}")

    return train_list, val_list, test_list


# ============================================================================
# 5. TRAINING LOOP
# ============================================================================

class Trainer:
    def __init__(self, model, train_loader, val_loader, criterion, optimizer,
                 scheduler, device, save_dir='checkpoints'):
        self.model = model
        self.train_loader = train_loader
        self.val_loader = val_loader
        self.criterion = criterion
        self.optimizer = optimizer
        self.scheduler = scheduler
        self.device = device
        self.save_dir = save_dir

        os.makedirs(save_dir, exist_ok=True)

        self.best_dice = 0.0
        self.best_iou = 0.0
        self.history = {
            'train_loss': [], 'val_loss': [],
            'train_iou': [], 'val_iou': [],
            'train_dice': [], 'val_dice': []
        }

    def train_epoch(self):
        self.model.train()
        running_loss = 0.0
        running_iou = 0.0
        running_dice = 0.0

        pbar = tqdm(self.train_loader, desc='Training')
        for images, masks in pbar:
            images = images.to(self.device)
            masks = masks.to(self.device)

            # Forward pass
            self.optimizer.zero_grad()
            outputs = self.model(images)
            loss = self.criterion(outputs, masks)

            # Backward pass
            loss.backward()
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)
            self.optimizer.step()

            # Calculate metrics
            with torch.no_grad():
                pred = torch.sigmoid(outputs)
                iou = calculate_iou(pred, masks)
                dice = calculate_dice(pred, masks)

            running_loss += loss.item()
            running_iou += iou
            running_dice += dice

            pbar.set_postfix({
                'loss': f'{loss.item():.4f}',
                'iou': f'{iou:.4f}',
                'dice': f'{dice:.4f}'
            })

        epoch_loss = running_loss / len(self.train_loader)
        epoch_iou = running_iou / len(self.train_loader)
        epoch_dice = running_dice / len(self.train_loader)

        return epoch_loss, epoch_iou, epoch_dice

    def validate_epoch(self):
        self.model.eval()
        running_loss = 0.0
        running_iou = 0.0
        running_dice = 0.0

        with torch.no_grad():
            pbar = tqdm(self.val_loader, desc='Validation')
            for images, masks in pbar:
                images = images.to(self.device)
                masks = masks.to(self.device)

                outputs = self.model(images)
                loss = self.criterion(outputs, masks)

                pred = torch.sigmoid(outputs)
                iou = calculate_iou(pred, masks)
                dice = calculate_dice(pred, masks)

                running_loss += loss.item()
                running_iou += iou
                running_dice += dice

                pbar.set_postfix({
                    'loss': f'{loss.item():.4f}',
                    'iou': f'{iou:.4f}',
                    'dice': f'{dice:.4f}'
                })

        epoch_loss = running_loss / len(self.val_loader)
        epoch_iou = running_iou / len(self.val_loader)
        epoch_dice = running_dice / len(self.val_loader)

        return epoch_loss, epoch_iou, epoch_dice

    def train(self, num_epochs, patience=15):
        """Complete training loop with early stopping"""
        no_improve = 0

        for epoch in range(num_epochs):
            print(f"\nEpoch {epoch+1}/{num_epochs}")
            print("-" * 50)

            # Train
            train_loss, train_iou, train_dice = self.train_epoch()

            # Validate
            val_loss, val_iou, val_dice = self.validate_epoch()

            # Update scheduler
            self.scheduler.step()

            # Save history
            self.history['train_loss'].append(train_loss)
            self.history['val_loss'].append(val_loss)
            self.history['train_iou'].append(train_iou)
            self.history['val_iou'].append(val_iou)
            self.history['train_dice'].append(train_dice)
            self.history['val_dice'].append(val_dice)

            # Print epoch summary
            print(f"\nTrain Loss: {train_loss:.4f} | IoU: {train_iou:.4f} | Dice: {train_dice:.4f}")
            print(f"Val Loss: {val_loss:.4f} | IoU: {val_iou:.4f} | Dice: {val_dice:.4f}")

            # Save best model
            if val_dice > self.best_dice:
                self.best_dice = val_dice
                self.best_iou = val_iou
                torch.save({
                    'epoch': epoch,
                    'model_state_dict': self.model.state_dict(),
                    'optimizer_state_dict': self.optimizer.state_dict(),
                    'dice': val_dice,
                    'iou': val_iou
                }, os.path.join(self.save_dir, 'best_model.pth'))
                print(f"✓ Model saved! Best Dice: {self.best_dice:.4f}, Best IoU: {self.best_iou:.4f}")
                no_improve = 0
            else:
                no_improve += 1

            # Early stopping
            if no_improve >= patience:
                print(f"\nEarly stopping triggered after {epoch+1} epochs")
                break

        print(f"\nTraining completed! Best Dice: {self.best_dice:.4f}, Best IoU: {self.best_iou:.4f}")

    def plot_history(self):
        """Plot training history - Loss, IoU, Dice curves"""
        fig, axes = plt.subplots(1, 3, figsize=(18, 5))

        epochs = range(1, len(self.history['train_loss']) + 1)

        # Plot 1: Loss
        axes[0].plot(epochs, self.history['train_loss'], 'b-', label='Train Loss', linewidth=2)
        axes[0].plot(epochs, self.history['val_loss'], 'r-', label='Val Loss', linewidth=2)
        axes[0].set_title('Training and Validation Loss', fontsize=14, fontweight='bold')
        axes[0].set_xlabel('Epoch', fontsize=12)
        axes[0].set_ylabel('Loss', fontsize=12)
        axes[0].legend(fontsize=11)
        axes[0].grid(True, alpha=0.3)

        # Plot 2: IoU
        axes[1].plot(epochs, self.history['train_iou'], 'b-', label='Train IoU', linewidth=2)
        axes[1].plot(epochs, self.history['val_iou'], 'r-', label='Val IoU', linewidth=2)
        axes[1].axhline(y=0.85, color='g', linestyle='--', label='Target (0.85)', linewidth=2)
        axes[1].set_title('Jaccard Index (IoU)', fontsize=14, fontweight='bold')
        axes[1].set_xlabel('Epoch', fontsize=12)
        axes[1].set_ylabel('IoU', fontsize=12)
        axes[1].legend(fontsize=11)
        axes[1].grid(True, alpha=0.3)

        # Plot 3: Dice
        axes[2].plot(epochs, self.history['train_dice'], 'b-', label='Train Dice', linewidth=2)
        axes[2].plot(epochs, self.history['val_dice'], 'r-', label='Val Dice', linewidth=2)
        axes[2].axhline(y=0.92, color='g', linestyle='--', label='Target (0.92)', linewidth=2)
        axes[2].set_title('Dice Coefficient', fontsize=14, fontweight='bold')
        axes[2].set_xlabel('Epoch', fontsize=12)
        axes[2].set_ylabel('Dice', fontsize=12)
        axes[2].legend(fontsize=11)
        axes[2].grid(True, alpha=0.3)

        plt.tight_layout()

        # Save plot
        save_path = os.path.join(self.save_dir, 'training_history.png')
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"\n✓ Training curves saved to: {save_path}")

        plt.show()


# ============================================================================
# 6. EVALUATION AND VISUALIZATION
# ============================================================================

def evaluate_model(model, test_loader, device):
    """Comprehensive evaluation on test set"""
    model.eval()
    all_metrics = []

    with torch.no_grad():
        for images, masks in tqdm(test_loader, desc='Evaluating'):
            images = images.to(device)
            masks = masks.to(device)

            outputs = model(images)
            pred = torch.sigmoid(outputs)

            # Calculate metrics for each image in batch
            for i in range(pred.shape[0]):
                metrics = calculate_metrics(pred[i], masks[i])
                all_metrics.append(metrics)

    # Aggregate metrics
    avg_metrics = {
        'iou': np.mean([m['iou'] for m in all_metrics]),
        'dice': np.mean([m['dice'] for m in all_metrics]),
        'sensitivity': np.mean([m['sensitivity'] for m in all_metrics]),
        'specificity': np.mean([m['specificity'] for m in all_metrics]),
        'precision': np.mean([m['precision'] for m in all_metrics])
    }

    print("\n" + "="*60)
    print("TEST SET EVALUATION RESULTS")
    print("="*60)
    for metric, value in avg_metrics.items():
        status = "✓ PASS" if (metric == 'iou' and value > 0.85) or (metric == 'dice' and value > 0.92) else ""
        print(f"{metric.capitalize():15s}: {value:.4f}  {status}")
    print("="*60)

    return avg_metrics


def visualize_predictions(model, dataset, device, num_samples=5, save_path='predictions.png'):
    """Visualize model predictions vs ground truth"""
    model.eval()
    fig, axes = plt.subplots(num_samples, 3, figsize=(15, 5*num_samples))

    # Handle case where we only want 1 sample
    if num_samples == 1:
        axes = axes.reshape(1, -1)

    indices = np.random.choice(len(dataset), num_samples, replace=False)

    with torch.no_grad():
        for idx, img_idx in enumerate(indices):
            image, mask = dataset[img_idx]

            # Predict
            image_tensor = image.unsqueeze(0).to(device)
            output = model(image_tensor)
            pred = torch.sigmoid(output).cpu().squeeze()

            # Denormalize image for visualization
            img_vis = image.permute(1, 2, 0).numpy()
            img_vis = img_vis * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
            img_vis = np.clip(img_vis, 0, 1)

            # Calculate metrics for this prediction
            pred_tensor = pred.unsqueeze(0).unsqueeze(0)
            mask_tensor = mask.unsqueeze(0)
            metrics = calculate_metrics(pred_tensor, mask_tensor)

            # Plot
            axes[idx, 0].imshow(img_vis)
            axes[idx, 0].set_title('Original Image', fontsize=12, fontweight='bold')
            axes[idx, 0].axis('off')

            axes[idx, 1].imshow(mask.squeeze(), cmap='gray')
            axes[idx, 1].set_title('Ground Truth Mask', fontsize=12, fontweight='bold')
            axes[idx, 1].axis('off')

            axes[idx, 2].imshow(pred > 0.5, cmap='gray')
            axes[idx, 2].set_title(f'Prediction\nIoU: {metrics["iou"]:.3f} | Dice: {metrics["dice"]:.3f}',
                                  fontsize=12, fontweight='bold')
            axes[idx, 2].axis('off')

    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    print(f"\n✓ Predictions visualization saved to: {save_path}")
    plt.show()


def save_sample_predictions(model, dataset, device, save_dir, num_samples=10):
    """Save individual prediction images for analysis"""
    model.eval()
    pred_dir = os.path.join(save_dir, 'sample_predictions')
    os.makedirs(pred_dir, exist_ok=True)

    indices = np.random.choice(len(dataset), num_samples, replace=False)

    with torch.no_grad():
        for i, img_idx in enumerate(tqdm(indices, desc='Saving predictions')):
            image, mask = dataset[img_idx]

            # Predict
            image_tensor = image.unsqueeze(0).to(device)
            output = model(image_tensor)
            pred = torch.sigmoid(output).cpu().squeeze().numpy()

            # Save as PNG
            pred_binary = (pred > 0.5).astype(np.uint8) * 255
            pred_img = Image.fromarray(pred_binary)
            pred_img.save(os.path.join(pred_dir, f'prediction_{i+1}.png'))

    print(f"✓ Saved {num_samples} sample predictions to: {pred_dir}")


# ============================================================================
# 7. MAIN EXECUTION PIPELINE
# ============================================================================

def main():
    """
    Complete end-to-end pipeline execution

    Steps:
    1. Data preparation and splitting
    2. Dataset and DataLoader creation
    3. Model initialization
    4. Training with early stopping
    5. Evaluation on test set
    6. Visualization of results
    """

    # ========================================================================
    # CONFIGURATION
    # ========================================================================
    CONFIG = {
        # Data paths - UPDATE THESE TO YOUR LOCAL PATHS
        'image_dir': '/content/drive/MyDrive/CMPE255/Project/ISIC2018_Task1-2_Training_Input',
        'mask_dir': '/content/drive/MyDrive/CMPE255/Project/ISIC2018_Task1_Training_GroundTruth',

        # Training hyperparameters (from project spec)
        'batch_size': 12,
        'num_epochs': 30,
        'learning_rate': 1e-3,
        'weight_decay': 1e-4,
        'patience': 10,  # Early stopping patience

        # Data
        'train_split': 0.8,
        'val_split': 0.1,
        'num_workers': 4,

        # Model
        'backbone': 'resnet50',  # or 'resnet101'
        'output_stride': 16,

        # Output
        'save_dir': 'checkpoints_deeplabv3plus',
        'seed': 42
    }

    # Set seeds
    set_seed(CONFIG['seed'])

    # Device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print("="*60)
    print("ISIC 2018 SKIN LESION SEGMENTATION - DeepLabV3+")
    print("="*60)
    print(f"Device: {device}")
    if torch.cuda.is_available():
        print(f"GPU: {torch.cuda.get_device_name(0)}")
        print(f"CUDA Version: {torch.version.cuda}")
    print("="*60)

    # ========================================================================
    # STEP 1: DATA PREPARATION
    # ========================================================================
    print("\n[STEP 1/7] Preparing data splits...")
    train_list, val_list, test_list = prepare_data_splits(
        CONFIG['image_dir'],
        train_split=CONFIG['train_split'],
        val_split=CONFIG['val_split']
    )

    # ========================================================================
    # STEP 2: CREATE DATASETS AND DATALOADERS
    # ========================================================================
    print("\n[STEP 2/7] Creating datasets and dataloaders...")

    train_dataset = ISICDataset(
        CONFIG['image_dir'],
        CONFIG['mask_dir'],
        train_list,
        transform=get_transforms('train')
    )
    val_dataset = ISICDataset(
        CONFIG['image_dir'],
        CONFIG['mask_dir'],
        val_list,
        transform=get_transforms('val')
    )
    test_dataset = ISICDataset(
        CONFIG['image_dir'],
        CONFIG['mask_dir'],
        test_list,
        transform=get_transforms('val')
    )

    train_loader = DataLoader(
        train_dataset,
        batch_size=CONFIG['batch_size'],
        shuffle=True,
        num_workers=CONFIG['num_workers'],
        pin_memory=True if torch.cuda.is_available() else False
    )
    val_loader = DataLoader(
        val_dataset,
        batch_size=CONFIG['batch_size'],
        shuffle=False,
        num_workers=CONFIG['num_workers'],
        pin_memory=True if torch.cuda.is_available() else False
    )
    test_loader = DataLoader(
        test_dataset,
        batch_size=CONFIG['batch_size'],
        shuffle=False,
        num_workers=CONFIG['num_workers'],
        pin_memory=True if torch.cuda.is_available() else False
    )

    print(f"✓ Train batches: {len(train_loader)}")
    print(f"✓ Val batches: {len(val_loader)}")
    print(f"✓ Test batches: {len(test_loader)}")

    # ========================================================================
    # STEP 3: MODEL INITIALIZATION
    # ========================================================================
    print("\n[STEP 3/7] Initializing DeepLabV3+ model...")

    model = DeepLabV3Plus(
        num_classes=1,
        backbone=CONFIG['backbone'],
        pretrained=True,
        output_stride=CONFIG['output_stride']
    ).to(device)

    num_params = sum(p.numel() for p in model.parameters()) / 1e6
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad) / 1e6

    print(f"✓ Model: DeepLabV3+ with {CONFIG['backbone']}")
    print(f"✓ Total parameters: {num_params:.2f}M")
    print(f"✓ Trainable parameters: {trainable_params:.2f}M")

    # ========================================================================
    # STEP 4: LOSS, OPTIMIZER, SCHEDULER
    # ========================================================================
    print("\n[STEP 4/7] Setting up loss, optimizer, and scheduler...")

    criterion = CombinedLoss(dice_weight=0.7, focal_weight=0.3)

    optimizer = optim.AdamW(
        model.parameters(),
        lr=CONFIG['learning_rate'],
        weight_decay=CONFIG['weight_decay']
    )

    scheduler = optim.lr_scheduler.OneCycleLR(
        optimizer,
        max_lr=CONFIG['learning_rate'],
        epochs=CONFIG['num_epochs'],
        steps_per_epoch=len(train_loader),
        pct_start=0.3
    )

    print(f"✓ Loss: Combined (Dice: 0.7 + Focal: 0.3)")
    print(f"✓ Optimizer: AdamW (lr={CONFIG['learning_rate']}, weight_decay={CONFIG['weight_decay']})")
    print(f"✓ Scheduler: OneCycleLR")

    # ========================================================================
    # STEP 5: TRAINING
    # ========================================================================
    print("\n[STEP 5/7] Starting training...")
    print(f"Target: IoU > 0.85, Dice > 0.92")
    print("-" * 60)

    trainer = Trainer(
        model, train_loader, val_loader,
        criterion, optimizer, scheduler,
        device, CONFIG['save_dir']
    )

    trainer.train(num_epochs=CONFIG['num_epochs'], patience=CONFIG['patience'])

    # Plot training history
    print("\n[STEP 5.5/7] Plotting training history...")
    trainer.plot_history()

    # ========================================================================
    # STEP 6: EVALUATION ON TEST SET
    # ========================================================================
    print("\n[STEP 6/7] Evaluating on test set...")

    # Load best model
    checkpoint = torch.load(os.path.join(CONFIG['save_dir'], 'best_model.pth'))
    model.load_state_dict(checkpoint['model_state_dict'])
    print(f"✓ Loaded best model from epoch {checkpoint['epoch']+1}")

    test_metrics = evaluate_model(model, test_loader, device)

    # ========================================================================
    # STEP 7: VISUALIZATION AND SAVING RESULTS
    # ========================================================================
    print("\n[STEP 7/7] Generating visualizations and saving results...")

    # Visualize predictions
    visualize_predictions(
        model, test_dataset, device,
        num_samples=5,
        save_path=os.path.join(CONFIG['save_dir'], 'predictions_visualization.png')
    )

    # Save sample predictions
    save_sample_predictions(
        model, test_dataset, device,
        CONFIG['save_dir'], num_samples=10
    )

    # Save configuration and results
    results = {
        'config': CONFIG,
        'training': {
            'best_val_dice': trainer.best_dice,
            'best_val_iou': trainer.best_iou,
            'total_epochs': len(trainer.history['train_loss'])
        },
        'test_metrics': test_metrics
    }

    with open(os.path.join(CONFIG['save_dir'], 'results.json'), 'w') as f:
        json.dump(results, f, indent=4)

    # ========================================================================
    # FINAL SUMMARY
    # ========================================================================
    print("\n" + "="*60)
    print("PIPELINE COMPLETED SUCCESSFULLY!")
    print("="*60)
    print("\nFINAL RESULTS:")
    print("-" * 60)
    print(f"Best Validation Dice:  {trainer.best_dice:.4f}")
    print(f"Best Validation IoU:   {trainer.best_iou:.4f}")
    print(f"Test Dice:             {test_metrics['dice']:.4f} {'✓ TARGET MET' if test_metrics['dice'] > 0.92 else ''}")
    print(f"Test IoU:              {test_metrics['iou']:.4f} {'✓ TARGET MET' if test_metrics['iou'] > 0.85 else ''}")
    print(f"Test Sensitivity:      {test_metrics['sensitivity']:.4f}")
    print(f"Test Specificity:      {test_metrics['specificity']:.4f}")
    print(f"Test Precision:        {test_metrics['precision']:.4f}")
    print("-" * 60)
    print(f"\nOUTPUT FILES (saved to {CONFIG['save_dir']}/):")
    print("  ✓ best_model.pth - Best model weights")
    print("  ✓ training_history.png - Loss/IoU/Dice curves")
    print("  ✓ predictions_visualization.png - Sample predictions")
    print("  ✓ sample_predictions/ - Individual prediction images")
    print("  ✓ results.json - Complete results and configuration")
    print("="*60)

    return model, test_metrics, trainer


# ============================================================================
# ENTRY POINT
# ============================================================================

if __name__ == "__main__":



    print("\n" + "="*60)
    print("Starting ISIC 2018 Skin Lesion Segmentation Pipeline")
    print("Model: DeepLabV3+ with ResNet50 Backbone")
    print("="*60 + "\n")

    model, test_metrics, trainer = main()

    print("\n✓ All done! Check the output directory for results.")
```
